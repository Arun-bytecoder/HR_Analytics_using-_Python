Summary Statistical: Value_counts() 

```python
data['Position'].value_counts()
```

```python
import seaborn as sns
```

```python
## visualize the HR position in each department and compare each department HR

plt.figure(figsize=(28,8))
sns.countplot(data = data, x = 'Position')
plt.xticks(rotation=45)
plt.show()
```

```python
## Count the Marital status for HR
data['MaritalStatus'].value_counts()
```

```python
# Visualize the count of HR marital status
data['MaritalStatus'].value_counts().plot.pie(autopct='%0.2f%%')
plt.show()
```

```python
# count the gender values 

data['Gender'].value_counts()
```

```python
# visualize the gender counts
sns.countplot(data=data, x="Gender",  palette=['green','blue'])
plt.show()
```

```python
## counts the employment status for HR 

data.EmploymentStatus.value_counts()
```

```python
# visualize the Employment status

sns.countplot(data=data, x="EmploymentStatus", palette=['blue','yellow','red'])
```

```python
## count the department

data.Department.value_counts()
```

```python
# visualize the department counts

data.Department.value_counts().plot.pie(autopct='%0.2f%%')
plt.show()
```

```python
data.RecruitmentSource.value_counts()
```

```python
data.RecruitmentSource.value_counts().plot.pie(autopct='%0.2f%%')
plt.show()
```

```python
data.PerformanceScore.value_counts()
```

```python
sns.countplot(data=data, x='PerformanceScore', palette=['blue','red','green','yellow'])
plt.show()
```
